**Color Quantization** is a process that reduces the number of distinct colors used in an image, usually with the intention that the new image should be as visually similar as possible to the original image. Color quantization is critical for displaying images with many colors on devices that can only display a limited number of colors, usually due to memory limitations, and enables efficient compression of certain types of images.
In this project, I used **KMeans Clustering** algorithm in order to reduce the number of used colors in a **parrot image**.
The final parrot image has 6 colors.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as implt
```

```python
# RGB CODES FOR EACH PIXEL
original_image_3darray=implt.imread('Parrot.jpg')
```

```python
# Show the original image
plt.figure(figsize=(6,6),dpi=200)
plt.imshow(original_image_3darray);
```

```python
height, width, color=original_image_3darray.shape
```

```python
(height, width, color)
```

```python
original_image_2darray=original_image_3darray.reshape(height*width,color)
```

```python
original_image_2darray.shape
```

```python
from sklearn.cluster import KMeans
```

```python
# clustering the collors into 6 clusters
kmeans_model=KMeans(n_clusters=6)
```

```python
kmeans_model.fit(original_image_2darray)
```

```python
# using cluster centers as 6 colors used in the new image
kmeans_model.cluster_centers_
```

```python
six_colors=kmeans_model.cluster_centers_.astype(int)
```

```python
six_colors
```

```python
labels=kmeans_model.labels_
```

```python
labels
```

```python
new_image_2darray=six_colors[labels]
```

```python
new_image_2darray
```

```python
new_image_3darray=new_image_2darray.reshape(height,width,color)
```

```python
# new RGB CODES FOR EACH PIXEL
new_image_3darray
```

```python
# Plotting the final image with 6 colors
plt.figure(figsize=(6,6),dpi=200)
plt.imshow(new_image_3darray)
plt.savefig('6 collor parrot.jpg');
```
